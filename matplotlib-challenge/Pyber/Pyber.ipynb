```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
city_data=pd.read_csv('/Users/zhongyuliu/Desktop/colnyc201907data3/Homework/05-Matplotlib/Instructions/Pyber/data/city_data.csv')
ride_data=pd.read_csv('/Users/zhongyuliu/Desktop/colnyc201907data3/Homework/05-Matplotlib/Instructions/Pyber/data/ride_data.csv')
```

```python
city=pd.DataFrame(city_data)
ride=pd.DataFrame(ride_data)
```

```python
city
```

```python
ride
```

```python
merged_data=pd.merge(city,ride,on='city')
```

```python
merged_data
```

```python
city_grouped=merged_data.groupby('city')

AverageFarePerCity=city_grouped['fare'].mean()
TotalNumberOfRidePerCity=city_grouped['ride_id'].count()
TotalNumberOfDrivesPerCity=city_grouped['driver_count'].mean()
CityType=pd.Series(city['type'].values,index =city['city'])
BubbleData=pd.DataFrame({
    'Average Fare per City':AverageFarePerCity,
    'Total Number of Ride per City':TotalNumberOfRidePerCity,
    'Total Number of Drivers per City':TotalNumberOfDrivesPerCity,
    'City Type':CityType
})
```

```python
BubbleData
```

```python
color_scheme = {'Gold':'#FFD700', 'Light Sky Blue':'#87CEFA', 'Light Coral':'#F08080'}
city_color = {'Urban': color_scheme['Gold'], 'Suburban': color_scheme['Light Sky Blue'], 'Rural': color_scheme['Light Coral']}
```

```python
x=BubbleData['Total Number of Ride per City']
y=BubbleData['Average Fare per City']
z=BubbleData['Total Number of Drivers per City']
t=BubbleData['City Type']

rural=BubbleData[BubbleData['City Type']=='Rural']
suburban=BubbleData[BubbleData['City Type']=='Suburban']
urban=BubbleData[BubbleData['City Type']=='Urban']

plt.scatter(rural['Total Number of Ride per City'], rural['Average Fare per City'], s = rural['Total Number of Drivers per City']*10, color = city_color['Rural'], edgecolor = 'black', label = 'Rural', alpha = .75)
plt.scatter(suburban['Total Number of Ride per City'], suburban['Average Fare per City'], s = suburban['Total Number of Drivers per City']*10, color = city_color['Suburban'], edgecolor = 'black', label = 'Suburban', alpha = .75)
plt.scatter(urban['Total Number of Ride per City'], urban['Average Fare per City'], s = urban['Total Number of Drivers per City']*10, color = city_color['Urban'], edgecolor = 'black', label = 'Urban', alpha = .75)

plt.xlabel('Total Number of Ride per City')
plt.ylabel('Average Fare per City')
plt.title('Pyber Ride Sharing Data(2016)')

lgnd = plt.legend(frameon = True, edgecolor = 'black')
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]

plt.show()
```

```python
Type=merged_data.groupby('type')['driver_count','type','fare','ride_id']
fare_sum = Type.sum()['fare']
labels = fare_sum.index
colors = [city_color[n] for n in labels]
explode = [0 , 0, .3]
plt.pie(fare_sum, startangle = 90, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = True, wedgeprops = {'linewidth': .5, 'edgecolor': 'black'})
plt.title('% of Total Fares by City Type')
plt.axis('equal')
plt.show()
```

```python
ride_sum = Type.count()['ride_id']

labels = ride_sum.index
plt.pie(ride_sum, startangle = 90, explode = explode, colors = colors, labels = labels, autopct = "%1.1f%%", shadow = True, wedgeprops = {'linewidth': .5, 'edgecolor': 'black'})
plt.title('% of Total Rides by City Type')
plt.axis('equal')
plt.show()
```

```python
driver_sum = city.groupby('type').sum()['driver_count']

labels = driver_sum.index
plt.pie(driver_sum, startangle = 125, explode = explode, colors = colors, labels = labels, autopct = "%1.1f%%", shadow = True, wedgeprops = {'linewidth': .5, 'edgecolor': 'black'})
plt.title('% of Total Drivers by City Type')
plt.axis('equal')
plt.show()
```
